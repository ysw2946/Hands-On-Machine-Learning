```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv")
df.head(4)
```

```python
df = df.drop("Unnamed: 0",axis=1)
df.shape
```

```python
dic = {
    'Unknown' : 'N',
    "Less than $40K" : "a",
    "$40K - $60K" : "b",
    "$60K - $80K": "c",
    "$80K - $120K" : "d",
    "$120K +" : "e",
}

df["newIncome"] = df.Income_Category.map(lambda x: dic[x])
df.newIncome.head(4)
```

```python
def changeIncome(x):
    if x == "Unknown":
        return "N"
    elif x == "Less than $40K":
        return "a"
    elif x == "$40K - $60K":
        return "b"
    elif x == "$60K - $80K":
        return "c"
    elif x == "$80K - $120K":
        return "d"
    elif x == "$120K +":
        return "e"
    
df["newIncome"] = df.Income_Category.apply(changeIncome)
df.newIncome.head()
```

```python
df["AgeState"] = df.Customer_Age.map(lambda x : x//10 *10)
df.AgeState.value_counts().sort_index()
```

```python
df.Customer_Age
```

```python
df["newEduLevel"] = df.Education_Level.map(lambda x : 1 if "Graduate" in x else 0)
df.newEduLevel.value_counts()
```

```python
df['newLimit'] = df.Credit_Limit.map(lambda x : 1 if x >= 4500 else 0)
df.newLimit.value_counts()
```

```python
def check(x):
    if x.Marital_Status == "Married" and x.Card_Category == "Platinum":
        return 1
    else:
        return 0
    
df["newState"] = df.apply(check,axis=1)
df.newState.value_counts()
```

```python
def sex(x):
    if x.Gender == "M" :
        return "male"
    else:
        return "female"
    
df["Gender"] = df.apply(sex,axis=1)
df.Gender.value_counts()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv")
df.head(4)
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]
    
    if hour == '24':
        hour = "00:00:00"
        
        finaldate = pd.to_datetime(date +" "+hour) + datetime.timedelta(days=1)
        
    else:
        hour = hour +":00:00"
        finaldate = pd.to_datetime(date + " "+hour)
    
    return finaldate
```

```python
df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.head()
```

```python
df["dayName"] = df["(년-월-일:시)"].dt.day_name()
df.dayName.head(4)
```

```python
df.groupby(["dayName","PM10등급"], as_index=False).size().head(4)
```
